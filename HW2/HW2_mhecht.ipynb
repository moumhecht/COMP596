## Problem 1: Analyze the following equations graphically. In each case, sketch the vector field on the real line, find all the fixed points and classify their stability.
### A.  $$\frac{dx}{dt}  = 2x^2-9$$

```python
n = 20

# inintialize
x = np.linspace(-3,3,n)
z = np.zeros(n)
zeros = np.zeros(n)
# fill array 
for ii in range(0,n):
    z[ii] = 2. * (x[ii])**2. - 9.

# plot
plt.plot(x,z)
plt.plot(x,zeros)
```

```python
x,t = np.meshgrid(np.linspace(-4,4,20),np.linspace(-10,10,10))
dx = 2.* x**2 - 9. 
dt = np.ones(np.size(t))
plt.quiver(x,t,dx,dt)
```

We can verify this mathematically, noting that the fixed points will be where $\frac{dx}{dt}=0$:
$$\frac{dx}{dt} = 2x^2 - 9 = 0 \implies \frac{9}{2}= x^2 \implies  x = \pm \sqrt{\frac{9}{2}} = \pm 2.121$$

Then the fixed point $x = -2.121$ goes from increasing to decreasing (from left to right) and is a stable fixed point.

The fixed point $x = 2.121$ goes from decreasing to increasing and is thus an unstable fixed point.

### B. $$\frac{dx}{dt} = 1-2\cos(x)$$

```python
n = 100

# inintialize
x = np.linspace(-10,10,n)
z = np.zeros(n)
zeros = np.zeros(n)
# fill array 
for ii in range(0,n):
    z[ii] = 1. - 2.* np.cos(x[ii])

# plot
plt.plot(x,z)
plt.plot(x,zeros)
```

```python
x,t = np.meshgrid(np.linspace(-6,8,10),np.linspace(-5,5,10))
dx =1.- 2.* np.cos(x)
dt = np.ones(np.size(t))
plt.quiver(x,t,dx,dt)
```

Setting $\frac{dx}{dt} = 0$, we get
$$\frac{dx}{dt} =1-2\cos(x) = 0 \implies \cos(x) = \frac{1}{2} \implies x = \cos^{-1}\left(\frac{1}{2}\right) $$
So $x = \frac{\pi}{3} +2\pi m$ or $x = \frac{5\pi}{3} + 2\pi m$ for some integer $m$.

For fixed points given by  $x = \frac{\pi}{3} +2\pi m$:

```python
np.pi/3. 
```

```python
np.pi/3 +2*np.pi
```

```python
np.pi/3 -2*np.pi
```

The fixed points go from decreasing to increasing and are unstable fixed points.

For fixed points given by $x = \frac{5\pi}{3} + 2\pi m$:

```python
5.* np.pi /3.
```

```python
5.* np.pi/3. - 2. * np.pi
```

The fixed points go from increasing to decreasing and are stable fixed points.
